```python
import pandas as pd
import numpy as np
import json
```

```python
import os
final_table_columns=['id','created_at','full_text','entities','truncated','in_reply_to_status_id','in_reply_to_user_id','user','retweeted_status','is_quote_status','retweeted_count','favorite_count','quoted_status_id','quoted_status_permalink','quoted_status','possibly_sensitive','extended_entities','withheld_in_countries']

cur_path = os.getcwd()
dataset_path = cur_path + "\dataset\\"
for file in os.listdir(dataset_path):
    filename = os.fsdecode(file)
    if filename.endswith(".json"):
        #  print(os.path.join(dataset_path, filename))
        data = pd.read_json(os.path.join(dataset_path, filename), lines=True)
        data.drop(columns=[col for col in data if col not in final_table_columns], inplace=True)
        data.to_csv(f'final_dataset/{filename.strip(".json")}.csv')
        continue
    else:
        continue
```
